```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="https://www.jumia.com.ng/catalog/?q=phones+with+128gb"
```

```python
HEADERS=({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/129.0.0.0 Safari/537.36','Accept-Language':'en-US, en;q=0.5'})
```

```python
webpage= requests.get(url, headers=HEADERS)
```

```python
webpage
```

```python
soup = BeautifulSoup(webpage.content,'html.parser')
```

```python
soup
```

```python
image_tag = soup.find('img', attrs={'data-src': True})
if image_tag:
    image = image_tag.get('data-src')  # Get the value of 'data-src' attribute
    print(image)
else:
    print("No image found.")
```

```python
soup.find("div",attrs={"class":"prc"}).text
```

```python
soup.find("div",attrs={"class":"stars _s"}).text
```

```python
ram_spec_div = soup.find('div', {'class': "attr-value"})
if ram_spec_div:
    ram_spec = ram_spec_div.text.strip()  # Extract and strip the text
    print(ram_spec)
else:
    print("No RAM specification found.")
```

```python
rom_spec_div = soup.find('div', {'class': "attr-value"})
if rom_spec_div:
    ram_spec = rom_spec_div.text.strip()  # Extract and strip the text
    print(rom_spec)
else:
    print("No ROM specification found.")
```

```python
titles=[]
prices=[]


#Find all product items on the page
for item in soup.find_all('article', attrs= {'class':'prd _fb col c-prd'}):
    #Extract the product title
    title=item.find("h3", attrs= {"class":"name"})
    titles.append(title.text.strip() )

    #Extract the product price
    price= item.find("div",{"class":"prc"})
    prices.append(price.text.strip() )

#Create a DataFrame using the extracted data
df= pd.DataFrame({
    'Title':titles,
    'Price': prices,

})

df.to_csv('jumia_phoneswith128gb.csv',index=False)
print("Data saved to jumia_phoneswith128gb.csv successfully!")
```
